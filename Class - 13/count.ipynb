```python
#Using While Loop - If Else

numbers = [12, 23, 45, 56, 67, 78, 89, 90]

Evens = 0
Odds = 0

i = 0
while i <= len(numbers) - 1:
    if numbers[i] % 2 == 0:
        Evens += 1
    else:
        Odds += 1
    i += 1

print('Even Numbers:', Evens)
print('Odd Numbers:', Odds)
```
